## Data Preprosessing

### Merge all datasets

#### 1. Join product_category and product_subcategory dataframes on ProductCategoryKey.
##### Essentially, we want the ProductCategoryKey from product_category to map on product_subcategory. The shape of the df should be (37,4)

```python
# Merge on 'ProductCategoryKey'
merged_df1 = pd.merge(prod_cat, prod_subcat, on='ProductCategoryKey', how='inner')
```

```python
merged_df1.head()
```

```python
merged_df1.shape
```

#### 2. Merge merged_df1 with product df on ProductSubcategoryKey. The shape of the df should be (293,14) since we are adding 3 columns from merged_df1 to the product dataframe.

```python
# Merge on 'ProductSubcategoryKey'
merged_df2 = pd.merge(prod, merged_df1, on='ProductSubcategoryKey', how='inner')
merged_df2.shape
```

```python
merged_df2.head()
```

#### 3. Merge merged_df2 with returns on ProductKey. The shape should be (1809,17) since we are adding 3 columns from returns to merged_df2

```python
# Merge on 'ProductSubcategoryKey'
merged_df3 = pd.merge(merged_df2, returns, on='ProductKey', how='inner')
merged_df3.shape
```

#### 4. Add 'Sales_2015', 'Sales_2016' and 'Sales_2017' as separate columns in Customer dataset, based on the 'CustomerKey'

```python
# Add a Year column to each sales dataset
sales_2015['Year'] = 2015
sales_2016['Year'] = 2016
sales_2017['Year'] = 2017
```

```python
# Merge customer dataset with sales_2015 dataset. The shape should be (2630, 21)
cust_2015 = pd.merge(customer, sales_2015, on='CustomerKey', how='inner')
```

```python
cust_2015.shape
```

```python
# Merge customer dataset with sales_2016 dataset. The shape should be (23935, 21)
cust_2016 = pd.merge(customer, sales_2016, on='CustomerKey', how='inner')
cust_2016.shape
```

```python
# Merge customer dataset with sales_2017 dataset. The shape should be (29481, 21)
cust_2017 = pd.merge(customer, sales_2017, on='CustomerKey', how='inner')
cust_2017.shape
```

```python
# Merge datasets on 'CustomerKey'
merged_df = cust_2015.merge(cust_2016, how='outer')
merged_df4 = merged_df.merge(cust_2017, how='outer')
merged_df4.shape
```

```python
merged_df4.head(3)
```

#### 5. Merge merged_df4 and merged_df3 on the common columns

```python
# Get the common columns between merged_df4 and merged_df3
common_columns = set(merged_df4.columns).intersection(merged_df3.columns)

# Print the common columns
print(common_columns)
```

```python
# Perform the merge based on the common columns
merged_df5 = pd.merge(merged_df4, merged_df3, on=list(common_columns), how='inner')
merged_df5.head()
```

```python
merged_df5.shape
```

```python
# To verify that there is no duplicates
duplicates = merged_df5[merged_df5.duplicated()]

# Display the duplicates
print(duplicates)
```

No duplicated rows

##### Handle 'Gender' null values

```python
# Display all null values of Prefix
gender_null = df[df['Prefix'].isna()]
gender_null.head(3)
```

Total of 113 customers don't have their gender record. However, we can compute them based on their First Names.

```python
gender_null['FirstName'].unique()
```

From the names above, we can list out all the male names.

```python
male = ['DARREN','ARTHUR','CHRISTOPHER','LUIS','GERALD','JSUS','NICOLAS','CHRISTIAN','TODD','JOSE','MIGUEL',
       'JACK','NATHANIEL','DOUGLAS','DEVIN','LOUIS','KEITH','BRANDON','FERNANDO','MICAH','CESAR','GAIL','REGINALD',
       'CRAIG','ZACHARY','DALTON','ALEJANDRO','JAMIE','LESLIE','RICARDO','SHAWN','BENJAMIN','LEVI','TERRY','JONATHAN',
       'MARCUS','HAROLD','IAN','JEROME','RANDALL','ORLANDO','EDWARD','ALEXANDER','THOMAS','DENNIS','RICKY','PEDRO',
       'EDDIE','BRETT','WYATT','BARRY','NOAH','ROBYN','JASON','NEIL','CEDRIC']

# Fill null values of gender based on the names listed

df.loc[df['Gender'].isnull(), 'Gender'] = df.loc[df['Gender'].isnull(), 'FirstName'].map(
    lambda x: 'M' if x in male else 'F'
    )

df.isnull().sum()
```

```python
# Check if the values are correctly replaced in Gender column
df[df['Prefix'].isnull()]
```

We can ignore the null values in 'ProductColor' and 'Prefix' since the columns are not part of the analysis

#### 2. Drop Unused Columns

```python
# Drop unneccesary columns
# We want to exclude ['Prefix','FirstName','LastName','EmailAddress','OrderNumber','ProductKey',
#'OrderLineItem','ProductSubcategoryKey','ProductSKU','ProductDescription','ProductCategoryKey', 'ProductColor']

df = df.drop(['Prefix','FirstName','LastName','EmailAddress','OrderNumber','ProductKey',
              'OrderLineItem','ProductSubcategoryKey','ProductSKU', 'ProductDescription','ProductCategoryKey','ProductColor'],
             axis=1)
df.shape
```

#### 3. Get customers' age from their birth date. And then, bin the age into groups

```python
df['BirthDate'].head()
```

```python
from datetime import datetime

# Convert BirthDate to datetime
df['BirthDate'] = pd.to_datetime(df['BirthDate'], format='%d-%m-%Y', errors='coerce')

# Check the converted column
print(df['BirthDate'].head())
```

```python
# Define today's date in 'now' variable
now = datetime.now()
now
```

```python
# Calucate customers' age from 'BirthDate'
df['Age'] = (now.year - df['BirthDate'].dt.year) - ((now.month - df['BirthDate'].dt.month) < 0)
df['Age'].head()
```

```python
# We can drop BirthDate column now
df = df.drop(columns=['BirthDate'])
df.head(3)
```

```python
age = np.sort(df['Age'].unique())
print(age)
```

```python
# Since we have customers whose age is more than 80, we will drop those customers as they are not our customer target group
# We need to know how many of them first
pd.set_option('display.max_rows', None)
count = df[df['Age'] > 80].shape[0]
print(count)
```

```python
# Drop rows where Age > 80
df = df[df['Age'] <= 80]
df.shape
```

```python
# Get min and max of age
print("Max of Age: ", df['Age'].max())
print("Min of Age: ", df['Age'].min())
print("Median of Age: ", df['Age'].median())
```

```python
# Define the age class based on age
bins = [40,50,60,70,80]
df['AgeClass'] = pd.cut(df['Age'], bins=bins)
```

```python
df['AgeClass'].value_counts()
```

#### 4. Convert all the date columns to datetime format

```python
# 'OrderDate'
df['OrderDate'] = pd.to_datetime(df['OrderDate'], format='%d-%m-%Y', errors='coerce')

# 'StockDate'
df['StockDate'] = pd.to_datetime(df['StockDate'], format='%d-%m-%Y', errors='coerce')

# 'Year'
df['Year'] = pd.to_datetime(df['Year'], format='%Y', errors='coerce')

# 'ReturnDate'
df['ReturnDate'] = pd.to_datetime(df['ReturnDate'], format='%d-%m-%Y', errors='coerce')
```

```python
# Check the format
print('OrderDate: ', df['OrderDate'].dtypes)
print('StockDate: ', df['StockDate'].dtypes)
print('Year: ', df['Year'].dtypes)
print('ReturnDate: ', df['ReturnDate'].dtypes)
```

#### 5. Bin Annual Income

```python
df['AnnualIncome'].dtypes
```

```python
# Remove dollar sign and commas, then convert to integer
df['AnnualIncome'] = df['AnnualIncome'].replace('[\$,]', '', regex=True).astype(int)
```

```python
df['AnnualIncome'].head()
```

```python
# Get min and max of AnnualIncome
print("Max of AnnualIncome: ", df['AnnualIncome'].max())
print("Min of AnnualIncome: ", df['AnnualIncome'].min())
print("Median of AnnualIncome: ", df['AnnualIncome'].median())
```

```python
# Define the annual income class
bins = [0, 10000, 40000, 80000, 120000, 160000, 200000]
df['AnnualIncomeClass'] = pd.cut(df['AnnualIncome'], bins=bins)
df.head()
```

```python
df['AnnualIncomeClass'].value_counts()
```

#### 6. Calculate Product Profit from ProductCost and ProductPrice

```python
df['ProductProfit'] = (df['ProductPrice'] - df['ProductCost']).round(2)
df.head(3)
```

#### 7. Convert OrderDate from YYYY-MM-DD to YYYY-MM

```python
# Format OrderDate as YYYY-MM using strftime
df['OrderDate'] = df['OrderDate'].dt.strftime('%Y-%m')

# Display the updated DataFrame
print(df[['OrderDate']].head())
```

#### 8. Again, drop unnecessary columns

```python
df = df.drop(['StockDate','TerritoryKey','ReturnDate', 'TotalChildren','HomeOwner','ProductName','ProductSize','ProductStyle','Age'], axis=1)
df.head(3)
```

```python
df.isnull().sum()
```

#### 9. Drop duplicated rows

```python
# Find and drop duplicated rows

duplicates = df[df.duplicated()]
duplicates.shape
```

```python
# Drop duplicates in place
df.drop_duplicates(inplace=True)
df.shape
```
